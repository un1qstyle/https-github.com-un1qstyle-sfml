```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
```

```python
#Определяем функцию, которая будет возвращать класс A-F в зависимости от баллов
def ClassByBall(lBall):
    Class = ''
    if lBall <= 10000:
        Class = 'A'
    elif lBall > 10000 and lBall <=100000:
        Class = 'B'
    elif lBall > 100000 and lBall <=200000:
        Class = 'C'
    elif lBall > 200000 and lBall <=400000:
        Class = 'D'
    elif lBall > 400000 and lBall <=700000:
        Class = 'E'
    else:
        Class = 'F'
    return Class
```

```python
#Добавление строкового столбца в датафрейм, с установкой значения в зависимости от LIMIT_BAL
df['Class'] = df['LIMIT_BAL'].map(ClassByBall)
#вывод первых двух строк для контроля
df.head(2)
```
